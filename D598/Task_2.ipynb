```python
# Load the D598 Data Set into a DataFrame
import pandas as pd
df = pd.read_excel("D598 Data Set.xlsx")  # No full path

```

```python
# Identify duplicate rows
duplicate_rows = df[df.duplicated()]
print("Duplicate Rows:\n", duplicate_rows)
```

```python
# Group records by "Business State" and calculate descriptive statistics
df_grouped = df.groupby("Business State").agg(
    Total_Long_Term_Debt_Mean=("Total Long-term Debt", "mean"),
    Total_Long_Term_Debt_Median=("Total Long-term Debt", "median"),
    Total_Long_Term_Debt_Min=("Total Long-term Debt", "min"),
    Total_Long_Term_Debt_Max=("Total Long-term Debt", "max"),

    Total_Equity_Mean=("Total Equity", "mean"),
    Total_Equity_Median=("Total Equity", "median"),
    Total_Equity_Min=("Total Equity", "min"),
    Total_Equity_Max=("Total Equity", "max"),

    Debt_to_Equity_Mean=("Debt to Equity", "mean"),
    Debt_to_Equity_Median=("Debt to Equity", "median"),
    Debt_to_Equity_Min=("Debt to Equity", "min"),
    Debt_to_Equity_Max=("Debt to Equity", "max"),

    Total_Liabilities_Mean=("Total Liabilities", "mean"),
    Total_Liabilities_Median=("Total Liabilities", "median"),
    Total_Liabilities_Min=("Total Liabilities", "min"),
    Total_Liabilities_Max=("Total Liabilities", "max"),

    Total_Revenue_Mean=("Total Revenue", "mean"),
    Total_Revenue_Median=("Total Revenue", "median"),
    Total_Revenue_Min=("Total Revenue", "min"),
    Total_Revenue_Max=("Total Revenue", "max"),

    Profit_Margin_Mean=("Profit Margin", "mean"),
    Profit_Margin_Median=("Profit Margin", "median"),
    Profit_Margin_Min=("Profit Margin", "min"),
    Profit_Margin_Max=("Profit Margin", "max")
)

# Display grouped summary statistics
print("\nGrouped Summary Statistics by Business State:\n", df_grouped)
```

```python
# Store grouped statistics output in a new DataFrame
df_grouped.to_csv("Grouped_Statistics.csv")
```

```python
# Filter records in the DataFrame by negative Debt-to-Equity values
negative_debt_to_equity = df[df["Debt to Equity"] < 0]
print("\nBusinesses with Negative Debt-to-Equity:\n", negative_debt_to_equity)
```

```python
# Create a new DataFrame for Debt-to-Income Ratio
debt_to_income = df[["Business ID", "Total Long-term Debt", "Total Revenue"]].copy()

# Define a function to safely calculate Debt-to-Income Ratio
def calculate_debt_to_income(debt, revenue):
    if revenue == 0:
        if debt == 0:
            return 0
        else:
            return 1
    else:
        return debt / revenue

# Apply the safe calculation function
debt_to_income["Debt to Income Ratio"] = debt_to_income.apply(
    lambda row: calculate_debt_to_income(row["Total Long-term Debt"], row["Total Revenue"]),
    axis=1
)

print("\nDebt-to-Income Data Frame:\n", debt_to_income)
```

```python
# Merge the Debt-to-Income DataFrame back into the original DataFrame
df_final = pd.merge(df, debt_to_income[["Business ID", "Debt to Income Ratio"]], on="Business ID", how="left")

# Display the final merged DataFrame
print("\nFinal Data Frame with Debt-to-Income Ratio:\n", df_final)
```
